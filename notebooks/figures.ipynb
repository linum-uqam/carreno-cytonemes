```python

%load_ext autoreload
%autoreload 2

import carreno.viz.vol as viz
import yaml
import os
import imageio as io
```

```python
# output location
output_folder = ""
dataset_folder = ""
path_to_config = ".."
with open(os.path.join(path_to_config, 'config.yml'), 'r') as file:
    infos = yaml.safe_load(file)
    output_folder = os.path.join(path_to_config, infos['BASE']['output'], 'figure')
    dataset_folder = os.path.join(path_to_config, infos['BASE']['dataset'])

if not os.path.exists(output_folder):
   os.makedirs(output_folder)
   
print("Output folder :", output_folder)
print("Dataset folder :", dataset_folder)
```
